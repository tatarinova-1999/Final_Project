# Использование модели в Jupyter Notebook

Пользователь вводит характеристики недвижимости, например:
- Площадь квартиры
- Количество комнат
- Количество этажей
- Близость к метро и другие значимые признаки (в соответствии с обучением модели)
## Запуск кода для загрузки модели и предсказания
В Jupyter Notebook загружается сохранённая модель, данные пользователя преобразуются в нужный формат, и производится вызов метода предсказания модели.
## Вывод результатов 
Пользователю выводится результат предсказания (прогнозируемая цена). 

```python
import joblib
import numpy as np
import warnings
warnings.filterwarnings("ignore")

model = joblib.load('RFR_model.joblib')

def get_user_input():
    # Ввод числовых признаков
    minutes_to_metro = float(input("Введите время до метро (в минутах): "))
    number_of_rooms = int(input("Введите количество комнат: "))
    area = float(input("Введите площадь квартиры (кв.м): "))
    living_area = float(input("Введите жилую площадь (кв.м): "))
    kitchen_area = float(input("Введите площадь кухни (кв.м): "))
    floor = int(input("Введите этаж: "))
    number_of_floors = int(input("Введите количество этажей в доме: "))
    
    # Ввод категориальных признаков через выбор опций
    print("Тип квартиры: 1 - New building (Новостройка), 0 - Secondary (Вторичка)")
    apartment_type_new_building = int(input("Введите 1 или 0: "))
    apartment_type_secondary = 1 - apartment_type_new_building  # Логическое дополнение
    
    print("Регион: 1 - Moscow, 0 - Moscow region")
    region_moscow = int(input("Введите 1 или 0: "))
    region_moscow_region = 1 - region_moscow
    
    print("Тип ремонта:\n 1 - Cosmetic (Косметический)\n 2 - Designer (Дизайнерский)\n 3 - European-style (В европейском стиле)\n 4 - Without renovation (Без ремонта)")
    renovation_input = int(input("Введите номер типа ремонта (0-4): "))
    renovation_cosmetic = 1 if renovation_input == 1 else 0
    renovation_designer = 1 if renovation_input == 2 else 0
    renovation_european = 1 if renovation_input == 3 else 0
    renovation_without = 1 if renovation_input == 4 else 0

    # Формируем массив 
    user_data = np.array([[ 
        minutes_to_metro,
        number_of_rooms,
        area,
        living_area,
        kitchen_area,
        floor,
        number_of_floors,
        apartment_type_new_building,
        apartment_type_secondary,
        region_moscow,
        region_moscow_region,
        renovation_cosmetic,
        renovation_designer,
        renovation_european,
        renovation_without
    ]])
    
    return user_data

if __name__ == "__main__":
    user_data = get_user_input()
    prediction = model.predict(user_data)
    print(f"Прогнозируемая цена недвижимости: {prediction[0]:,.2f}".replace(',', ' ') + " рублей")
```
